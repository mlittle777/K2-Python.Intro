```python
import matplotlib
import matplotlib.pyplot as plt
```

```python
x = [-1, 3, 4, 8 , 10]
f = [-1, -2, 7, 13 , 1]
```

```python
%matplotlib inline
```

```python
# Plot data with a line and dots at the data points (this is what '-o' does)
plt.plot(x, f, '-o')  

# Attach labels and title (using LaTeX syntax)
plt.xlabel('$x$')
plt.ylabel('$f$')
plt.title("Simple plot of $x$ vs $f$");
```
